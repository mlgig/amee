```python
df = pd.read_csv('filtered.csv')
ds_list = df['dataset'].tolist()
# col_names = df.columns.values.tolist()
# col_names = col_names.append(['best_method', 'worst_method'])
best, worst = [],[]

for ds in ds_list:
    if ds not in ['Car','Wafer']:
        print(ds)
        path = glob.glob('./output/%s_*.csv'%(ds))
        if not path:
            print('No comparison result exists')
            print('---')

        else:
            path = path[0]
            auc_df =pd.read_csv(path)
        #     summarize_auc(auc_df)
            x = get_final_ranking(auc_df,beautify_display=False)
            curr_best = x['XAI_method'][x['scaled_ranking'].idxmin()] 
            curr_worst = x['XAI_method'][x['scaled_ranking'].idxmax()] 
            print('Best method: ', curr_best)
            print('Worst method: ', curr_worst)
            print('---')

```
